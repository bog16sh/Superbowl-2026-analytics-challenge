```python
df_unique = df.drop_duplicates(subset="id", keep="first")
```

```python
import re
from collections import Counter

# Function to extract hashtags
def extract_hashtags(text):
    if isinstance(text, str):
        return re.findall(r'#\w+', text)
    return []

# Add hashtags column
df_unique["hashtags"] = df_unique["text"].apply(extract_hashtags)

# Flatten hashtags
all_hashtags = [h for sublist in df_unique["hashtags"] for h in sublist]

unique_hashtags = set(all_hashtags)

#print("Unique Hashtags:")
for hashtag in sorted(unique_hashtags):
    print(hashtag)

#print(f"\nTotal unique hashtags found: {len(unique_hashtags)}")

# Count frequencies
hashtag_counts = Counter(all_hashtags)

print("\nTop 20 Hashtags:")
for hashtag, count in hashtag_counts.most_common(20):
    print(f"{hashtag}: {count}")

# Filter specific hashtag
ultrarace_tweets = df_unique[df_unique["text"].str.contains("#ULTRARace", na=False)]

print(f"\nFound {len(ultrarace_tweets)} rows with '#ULTRARace'")
print(ultrarace_tweets["text"].head().tolist())
```

```python
import re

# Dictionary mapping abbreviations to full names
state_dict = {
    "AL": "Alabama", "AK": "Alaska", "AZ": "Arizona", "AR": "Arkansas",
    "CA": "California", "CO": "Colorado", "CT": "Connecticut",
    "DE": "Delaware", "FL": "Florida", "GA": "Georgia",
    "HI": "Hawaii", "IA": "Iowa", "ID": "Idaho",
    "IL": "Illinois", "IN": "Indiana", "KS": "Kansas",
    "KY": "Kentucky", "LA": "Louisiana", "MA": "Massachusetts",
    "MD": "Maryland", "ME": "Maine", "MI": "Michigan",
    "MN": "Minnesota", "MO": "Missouri", "MS": "Mississippi",
    "MT": "Montana", "NC": "North Carolina", "ND": "North Dakota",
    "NE": "Nebraska", "NH": "New Hampshire", "NJ": "New Jersey",
    "NM": "New Mexico", "NV": "Nevada", "NY": "New York",
    "OH": "Ohio", "OK": "Oklahoma", "OR": "Oregon",
    "PA": "Pennsylvania", "RI": "Rhode Island",
    "SC": "South Carolina", "SD": "South Dakota",
    "TN": "Tennessee", "TX": "Texas", "UT": "Utah",
    "VA": "Virginia", "VT": "Vermont",
    "WA": "Washington", "WI": "Wisconsin",
    "WV": "West Virginia", "WY": "Wyoming"
}

df_unique["location"] = df_unique["location"].fillna("").str.upper()

def extract_state(location):
    if not isinstance(location, str):
        return None

    location_upper = location.upper()

    # First check full state names
    for full_name in state_dict.values():
        if full_name.upper() in location_upper:
            return full_name

    # Then check abbreviations
    for abbr, full_name in state_dict.items():
        if re.search(rf'\b{abbr}\b', location_upper):
            return full_name

    return None

df_unique["state"] = df_unique["location"].apply(extract_state)

df_unique["state"].value_counts().head(20)
```

```python
# We will keep it just in case, only 35% of missings. Going to be in the different dataset

df_annotations = df_unique[df_unique["entities.annotations"].notna()]
```

```python
# Tweets with URl has more attractions or not. Binary

df_unique["has_url"] = df_unique["entities.urls"].notna().astype(int)
```

```python
# Whether tweet is reply / retweet / quote. Binary too

df_unique["is_reply_or_rt"] = df_unique["referenced_tweets"].notna().astype(int)
```

```python
#Dropping these columns since 80-100# content of them is missing, and edit_controls., conversation_id and reply_settings are not relevant to hashtags analysis

df_unique = df_unique.drop(columns=[
    "edit_history_tweet_ids",
    "reply_settings",
    "entities.urls",
    "entities.annotations",
    "attachments.media_keys",
    "attachments.media_source_tweet_id",
    "edit_controls.edits_remaining",
    "edit_controls.is_edit_eligible",
    "edit_controls.editable_until",
    "referenced_tweets",
    "entities.mentions",
    "in_reply_to_user_id",
    "entities.hashtags",
    "geo.place_id",
    "entities.cashtags",
    "withheld.country_codes",
    "attachments.poll_ids",
    "name",
    "conversation_id"
])
```

```python

# 1️⃣ Convert created_at to datetime
df_unique["created_at"] = pd.to_datetime(df_unique["created_at"])

# 2️⃣ Create total engagement metric
df_unique["engagement_total"] = (
    df_unique["public_metrics.like_count"] +
    df_unique["public_metrics.retweet_count"] +
    df_unique["public_metrics.reply_count"] +
    df_unique["public_metrics.quote_count"]
)

# 3️⃣ Optional: Create engagement rate (very useful)
df_unique["engagement_rate"] = (
    df_unique["engagement_total"] /
    df_unique["public_metrics.impression_count"]
)

# Replace infinite values if impression_count was 0
df_unique["engagement_rate"] = df_unique["engagement_rate"].replace([float("inf")], 0)



df_unique.info()
```

Data Preparation Summary


1) Removed duplicate tweets
Since the dataset contained duplicate records, we removed all duplicate rows based on the unique id to ensure each tweet appears only once.

2) Dropped irrelevant and highly missing columns
We removed columns with 80–100% missing values, as well as columns not relevant for keyword analysis (e.g., edit_controls.*, conversation_id, and reply_settings) to improve data quality and reduce noise.

3) Handled missing location values
All missing values in the location column were replaced with "Unknown" to allow consistent geographic analysis as well as removed messy data and keep only states and some countries.

4) Created binary variable for URLs (has_url)
The entities.urls column was transformed into a binary variable (has_url), where:
1 = tweet contains a URL
0 = tweet does not contain a URL
This allows us to analyze whether tweets containing URLs generate higher engagement.

5) Created binary variable for replies/retweets
The referenced_tweets column was converted into a binary indicator to identify whether a tweet is a reply/retweet or an original post.

6) Prepared date for time analysis
The created_at column was converted to datetime format to enable time-based analysis and visualization in Tableau.

7) Created engagement metrics
We created:
engagement_total: the sum of likes, retweets, replies, and quote tweets to represent total interaction.
engagement_rate: engagement relative to impressions, to measure performance efficiency.
